```python
from tiny_model import *

device = 'cuda:0'
model = TinyModel().to(device)
```

```python
import datasets
import torch

dataset = datasets.load_dataset('noanabeshima/TinyStoriesV2', split='validation')
dataloader = torch.utils.data.DataLoader(dataset, batch_size=100)
```

```python
for batch in dataloader:
    batch, attn_mask = enc(batch['text'], max_length=128, return_attn_mask=True)
    with torch.no_grad():
        batch = batch[attn_mask[:,-1] == 1].to(device)
        acts = model['A2.0'](batch)
    break
```
